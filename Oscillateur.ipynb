```python
import numpy as np
from qutip import *
import matplotlib.pyplot as plt
width=4.2679
```

# Oscillateur

```python
N=30
a=destroy(N)
omega=1
H=omega*(a.dag()*a+1/2)
```

```python
psi0=fock(N,3)
fig,ax=plt.subplots()
plot_wigner(psi0,fig,ax)
ax.set_title("")
ax.grid()
fig.set_size_inches((width/2,width/2))
fig.tight_layout()
fig.set_dpi(300)
fig.savefig("pres_wigner_fock.png")
```

```python
psi0=coherent(N,2*np.exp(1j*1))
fig,ax=plt.subplots()
AR=3/4
plot_wigner(psi0,fig,ax)
ax.set_title("")
fig.set_size_inches((width/2,width/2))
ax.grid()
fig.tight_layout()
fig.set_dpi(300)
fig.savefig("pres_wigner_coh.png")
```

```python
psi0=coherent(N,2*np.exp(1j*1))+coherent(N,-2*np.exp(1j*1))
fig,ax=plt.subplots(1,2)
AR=3/4
plot_wigner_fock_distribution(psi0,fig,ax)
fig.set_size_inches((width,width/2))
ax[0].set_ylabel("")
ax[0].set_xlabel("")
ax[1].set_xlabel("")
ax[1].set_ylabel("")
ax[1].set_title("")
ax[0].grid()
fig.tight_layout()
fig.set_dpi(300)
fig.savefig("pres_wigner_cat.png")
```

```python
psi0=coherent_dm(N,2*np.exp(1j*1))+coherent_dm(N,-2*np.exp(1j*1))
psi1=coherent(N,2*np.exp(1j*1))+coherent(N,-2*np.exp(1j*1))

fig,ax=plt.subplots(1,2)
AR=3/4
plot_wigner(psi1,fig,ax[0])
plot_wigner(psi0,fig,ax[1])
fig.set_size_inches((width,width/2))
ax[0].set_ylabel("")
ax[0].set_xlabel("")
ax[1].set_xlabel("")
ax[1].set_ylabel("")
ax[1].set_title("Classique")
ax[0].set_title("Superoposition")
ax[0].grid()
ax[1].grid()

fig.tight_layout()
fig.set_dpi(300)
fig.savefig("pres_wigner_cat_dec.png")
```
