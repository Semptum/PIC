```python
import numpy as np
from qutip import *
import matplotlib.pyplot as plt
width=4.2679
```

# Qubit

```python
sx,sy,sz=sigmax(),sigmay(),sigmaz()
E0=1
E1=2
H=E0*(1+sz)/2+E1*(1-sz)/2
g,e=fock(2,0),fock(2,1)
```

```python
theta=1/2
phi=1
psi0=np.cos(theta/2)*g+np.sin(theta/2)*np.exp(1j*phi)*e
sph=Bloch()
sph.add_states(psi0)
sph.show()
```

```python
TIMES=np.linspace(0,10,500)
sol=mesolve(H,psi0,TIMES,[],[sx,sy,sz])
```

```python
fig,ax=plt.subplots()
AR=3/4
fig.set_size_inches((width,width*AR))
ax.plot(sol.times,sol.expect[0],label=r"$\sigma_x$")
ax.plot(sol.times,sol.expect[1],label=r"$\sigma_y$")
ax.plot(sol.times,sol.expect[2],label=r"$\sigma_z$")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_xlabel("Temps")
ax.set_ylabel("Coordonne")
fig.tight_layout()
fig.set_dpi(300)
fig.savefig("pres_rabi.png")
```
